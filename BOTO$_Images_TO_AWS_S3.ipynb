```python
import boto3
import os

# Initialize S3 client
s3_client = boto3.client(
    's3',
    aws_access_key_id='ASIAS7BGNQSZ6VAIH4JK',
    aws_secret_access_key='looFpwg7TtF3GPyoZxonlu9PspND775wlHyed3RN',
    aws_session_token='IQoJb3JpZ2luX2VjECEaCXVzLXdlc3QtMiJHMEUCIB0Mn0rIc2J8NDZNElmd9cB7J1aTkEr4DxoXBV6/cFVNAiEA+/zKRSAhM7AB30IehIVsSy0iA0+kKSurqbFZ14hwRJYqsQIIyv//////////ARAAGgwyMDQwOTE1MjQyNzUiDFd41hqax6Fu74bExSqFAgQOfbov7NhGM30x9abWw9IcLJ0HA2lWYRWGvU61cLphg3Y80VC31kJKRUlXid0jg0qx39PGLHs8OypSf52CCzlKPo7q1o1vX9QP0vyRlN9V1+u0Rop5Aa25V2IDO0Vewdsuz5WxHKTVjwAY67MtHL79naTSp3O6rPG7EQfEKmyDfvUSWv+ij7uirta+jmc4o8ESIq+mkLGEcn1UyzCpw0kO0YbLbxObaogWDHdWtK6Hr8kmB10Eqr6cY1UrDYYTh9TG2NBdJZGWLRtnKL3EkKegEPliZeEPcyd3XOUUK+AnB2Emg5wcP03qw9EQSg/QdB6IuFcwNYUvV52w4stBXA2FAHo4FDDUorm6BjqdAcx1FT0w5Pwkjx7d0nzE/5qVGc7kSipV5ijvuX0v93mlvM5qhNpL7eq1UhNRb8IGcoNU/OQWpMn1J/gUByjL8VqDS+zIJtvgJv4JCzGQ0pcroZxKDyXh6h+oN4Z011MeqHwxMLQyCHts0KOF5fmGpTRhMJ9r0MyFmLsyXPtvxZ2bJelHXeJ6MEfSAa6mw1C0PnAolR+NGxq028CkVfs='
)

# Create bucket (if not already created)
bucket_name = 'cancerdetection-datasource'
try:
    s3_client.create_bucket(Bucket=bucket_name)
    print(f"Bucket {bucket_name} created successfully.")
except s3_client.exceptions.BucketAlreadyOwnedByYou:
    print(f"Bucket {bucket_name} already exists.")

# Define local training set directory
logdir_train = "/Users/dhruvalokesh/Downloads/train"  
logdir_test = "/Users/dhruvalokesh/Downloads/test"    
train_sample_size = 2500
test_sample_size = 800

# Function to upload a limited number of files to S3
def upload_limited_files(local_folder, s3_bucket, s3_folder, sample_size):
    uploaded_files = 0
    for root, dirs, files in os.walk(local_folder):
        for file in files:
            if uploaded_files >= sample_size:
                break
            local_path = os.path.join(root, file)
            s3_path = os.path.join(s3_folder, os.path.relpath(local_path, local_folder))
            s3_client.upload_file(local_path, s3_bucket, s3_path)
            print(f"Uploaded {local_path} to s3://{s3_bucket}/{s3_path}")
            uploaded_files += 1
        if uploaded_files >= sample_size:
            break

# Upload 2500 training images to S3 under 'trainset' folder
upload_limited_files(logdir_train, bucket_name, 'trainset', train_sample_size)

# Upload 800 test images to S3 under 'testset' folder
upload_limited_files(logdir_test, bucket_name, 'testset', test_sample_size)

print("Training and test sets uploaded successfully!")
```
